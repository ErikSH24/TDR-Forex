```python
data = load_data("EURUSD","M1")
data['EMA200'] = talib.EMA(data['close'], timeperiod=200*60)
data['EMA30'] = talib.EMA(data['close'], timeperiod=30)
data = data.dropna().reset_index(drop = True)

cond_long = (data.EMA30 > data.EMA200) & (data.EMA30.shift(1) < data.EMA200.shift(1))
cond_short = (data.EMA30 < data.EMA200) & (data.EMA30.shift(1) > data.EMA200.shift(1))
data["cond_open_pos"] = cond_long * 1 - cond_short * 1
data["cond_close_pos"] = cond_short * 1 - cond_long * 1

data.head()
```

```python
signals = data[data.cond_open_pos != 0].reset_index(drop = True)
signals.head()
```

```python
dicti = {
    "dynamic_lot_size": False, 
    "equity_percent_loss": 15,
    "fixed_lot_size": 0.01, 
    "max_open_trades": 1, 
    "max_minute_open_trade": None, 
    "min_between_trades": None
}
symbol = "EURUSD"
data_H1 = load_data(symbol, "H1")
data = load_data(symbol,"M1")

data_H1['EMA200'] = talib.EMA(data_H1['close'], timeperiod=200*60)
data_H1['EMA30'] = talib.EMA(data_H1['close'], timeperiod=30)
data_H1 = data_H1.dropna().reset_index(drop = True)

cond_long = (data_H1.EMA30 > data_H1.EMA200) & (data_H1.EMA30.shift(1) < data_H1.EMA200.shift(1))
cond_short = (data_H1.EMA30 < data_H1.EMA200) & (data_H1.EMA30.shift(1) > data_H1.EMA200.shift(1))
data_H1["signal"] = cond_long * 1 - cond_short * 1
data_H1["cond_close_long"] = 0
data_H1["cond_close_short"] = 0
data_H1["sl_long"] = 50
data_H1["sl_short"] = 50
data_H1["tp_long"] = 50
data_H1["tp_short"] = 50

data["n_minute"] = data.groupby(["year", "month", "day", "hour"]).time.cumcount()+1
cols = ['year', 'day', 'month', 'hour', 'EMA200', 'EMA30',
       'signal', 'cond_close_long', 'cond_close_short', 'sl_long', 'sl_short',
       'tp_long', 'tp_short']
data = data.merge(data_H1[cols], on = ['year', 'day', 'month', 'hour'], how = "inner")
data.loc[data.n_minute > 1, "signal"] = 0

print(f"Operacions llargues: {sum(data.signal == 1)}")
print(f"Operacions curtes: {sum(data.signal == -1)}")

data.head()
```
